### Model using Linear Regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
df=pd.read_csv(r"C:\Users\HP\Desktop\ml\solubility.csv")
```

```python
df
```

```python
## data sepration
x=df.drop('logS',axis=1)
y=df['logS']
```

```python
x
y
```

```python
# training and testing splitting
x_train , x_test , y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=100)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_lr_train_prediction = lr.predict(x_train)
y_lr_test_prediction=lr.predict(x_test)
```

```python
print(y_lr_train_prediction)
```

```python
#20% of the data 
y_lr_test_prediction
```

```python
#model performance by comparing test and train
from sklearn.metrics import mean_squared_error,r2_score

## for training data 
lr_train_mean= mean_squared_error(y_train , y_lr_train_prediction)
#correlation coefficients
lr_train_r2= r2_score(y_train, y_lr_train_prediction)

## for testing data 
lr_test_mean= mean_squared_error(y_test , y_lr_test_prediction)
#correlation coefficients
lr_test_r2= r2_score(y_test, y_lr_test_prediction)
```

```python
lr_train_mean
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mean, lr_train_r2, lr_test_mean, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```
